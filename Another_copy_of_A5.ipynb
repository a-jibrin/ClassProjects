```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('IMDB Dataset.csv')

print(f"Number Rows:", data.shape[0])
print(f"Number Columns:", data.shape[1])
```

```python
sample_data = data.loc[data.index.isin(data.sample(n=20000, random_state=0).index)]
sample_data
```

#Text Preprocessing [10]
- Convert all reviews to lower case.
- Remove non-alphabetic words or tokens from the reviews.

```python
data['cleaned_reviews'] = data.review.str.lower()
data.head()
```

```python
def remove_non_alphabetic_words(string):
    words = string.split()
    words = [w for w in words if w.isalpha() == True]
    return ' '.join(words)

data['cleaned_reviews'] = data.cleaned_reviews.apply(remove_non_alphabetic_words)
data.head()
```

#Feature Extraction [10]
Use TfiDfVectorizer to convert the reviews to feature vectors. Use 10K most frequent words/tokens. Remove stopwords.

```python
import nltk
nltk.download('stopwords')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords

vectorizer = TfidfVectorizer(max_features=10000,
                             stop_words=stopwords.words('english'))
X = vectorizer.fit_transform(data.cleaned_reviews)
```

# Train Test Split [5]
- Split the data into train and test with a 75:25 ratio.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, data.sentiment, test_size=0.25, random_state=0)
print("Train Size:", X_train.shape[0])
print("Test Size:", X_test.shape[0])
```

#Algorithm Selection [20]
- Use Decisiion Trees, K Nearest Neighbors, and Support Vector Classifier to build classification models.
- Which performs the best in terms of precision and recall? Print classification reports for each of the above algorithms.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

dtclf = DecisionTreeClassifier().fit(X_train, y_train)
dt_pred = dtclf.predict(X_test)

print(classification_report(y_test, dt_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
# K Nearest Neighbors
knnclf = KNeighborsClassifier().fit(X_train, y_train)

knn_pred = knnclf.predict(X_test)
print(classification_report(y_test, knn_pred))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
# Support Vector Classifier
svcclf = SVC().fit(X_train, y_train)
svc_pred = svcclf.predict(X_test)
print(classification_report(y_test, svc_pred))
```
